```python

class Pendulo1_invertido:
    def __init__(self,m, MP, l, g, f, b, I, fza):
        self.m = m
        self.MP = MP
        self.l = l
        self.g = g
        self.f = f                #lo que en las ecuaciones es c
        self.b = b   #lo que en las ecuaciones es b
        self.I = I
        self.fza = fza

       
        #self.masa_tot = Masa_pendulo + masa_carro
        #self.denominador = Masa_pendulo*masa_carro*longitud**2+Inercia*masa_tot
        #self.num_xbiprima = masa_carro*longitud**2 + Inercia
        #self.masa_longitud = masa_carro*longitud

    def x(self):
        return self.trajectory[:,0]

    def xprima(self):
        return self.trajectory[:,1]

    def theta(self):
        return self.trajectory[:,2]

    def thetap(self):
        return self.trajectory[:,3]

    def plot(self):
        fig, ax = plt.subplots(5,1, figsize = (10,8), sharex=False)

        ax[0].plot(self.tau, self.x(), label="x", color="blue")
        ax[1].plot(self.tau, self.xprima(), label="x prima", color="green")  
        ax[2].plot(self.tau, self.theta(), label="theta", color="blue")
        ax[3].plot(self.tau, self.thetap(), label=" theta prima", color="green")
        ax[4].plot(self.xprima(), self.thetap(), label=" x ", color="green")



        ax[0].set_ylabel("x (m)")
        ax[0].set_xlabel("tiempo (s)")

        ax[1].set_ylabel("x prima (m/s)")
        ax[1].set_xlabel("tiempo (s)")

        ax[2].set_ylabel("theta (radianes)")
        ax[2].set_xlabel("tiempo (s)")

        ax[3].set_ylabel("theta prima (rad/s)")
        ax[3].set_xlabel("tiempo (s)")
                         
        ax[4].set_ylabel("theta (rad)")
        ax[4].set_xlabel("x (m)")
        
    def inicial_condicion(self, x_i, xprima_i,theta_i,thetap_i):
        self.x_i = x_i
        self.xprima_i = xprima_i
        self.theta_i = theta_i
        self.thetap_i =thetap_i

    def dynamics (self, state, t):

        g0 = state[1]
        g1 = (self.fza*(self.MP*self.l**2+self.I)-self.f*(self.MP*self.l**2+self.I)*state[1]-self.MP*self.l*self.b*state[3]+self.MP**2*self.g*self.l**2*state[2])/(self.I*(self.m+self.MP)+self.m*self.MP*self.l**2)
        g2 = state[3]
        g3 = (self.fza*self.MP*self.l-self.f*self.MP*self.l*state[1]-self.b*state[3]*(self.m+self.MP)+self.MP*self.g*self.l*state[2]*(self.MP+self.m))/(self.I*(self.m+self.MP)+self.m*self.MP*self.l**2)
        return np.array ([g0, g1, g2 ,g3])

    def integrate(self, num_steps, t_i, t_f, metodo):
        self.tau, self.dt = np.linspace(t_i, t_f, num=num_steps, retstep=True)
        self.trajectory = np.zeros([num_steps, 4])
        self.trajectory[0,0] = self.x_i
        self.trajectory[0,1] = self.xprima_i
        self.trajectory[0,2] = self.theta_i
        self.trajectory[0,3] = self.thetap_i

        for j in range (num_steps -1):
            self.trajectory[j+1] = metodo(self.trajectory[j], self.tau[j], self.dt, self.dynamics)
```

```python
def RK2(y, t, dt, derivadas):
    k0=dt*derivadas(y,t)
    k1=dt*derivadas(y+k0, t+dt)
    y_next=y+0.5*(k0+k1)
    return y_next
```

```python
def interactivo(m, MP, l, g, f,b, I, fza):
    p4=Pendulo1_invertido(m, MP, l, g, f,b, I, fza)
    p4.inicial_condicion(x_i=0, xprima_i=0.00, theta_i=np.pi, thetap_i=0)
    #p4.inicial_condicion(x_i=0, xprima_i=0.00, theta_i=pi, thetap_i=0)
    N=1000
    tiempo_ini=0
    tiempo_fin=10
    p4.integrate(N, tiempo_ini, tiempo_fin, RK2)
    p4.plot()
```
