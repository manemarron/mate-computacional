(e) Encapsula lo que acabamos de hacer en una función llamada criba_eratostenes, que reciba como argumento maximo el cual representa el tamaño del array. Muestra el resultado para maximo = {10, 100, 200}, usando un for

```python
def criba_eratostenes(m):
    b_arr= np.ones((m,), dtype=bool)
    n_max = int(np.sqrt(len(b_arr)))
    
    for j in range(2, n_max):
        
        b_arr[2*j::j] = False
        
    return np.nonzero(b_arr)
    
```

# Pregunta 4

El Sudoku es un pasatiempo que consiste en un arreglo 9×9 de celdas, con dígitos del 1 al 9, de tal manera que no se repiten valores en cada columna, cada renglón y en cada subarreglo de 3×3.

(a) Crea una función llamada correcto que reciba un array 1d de tamaño 9 e indique si está completo o no (i.e. incluye los números del 1 al 9).

```python
def correcto(A):
    total=0
    for n in range (9):
        if A[n]<=9:
            for m in range(1,10):
                
                if A[n]==m:
                    total= total+1
                    if total==9:
                        return "Completo"
        else:
            return "No completo"
```

```python
Ar= (1+np.arange(9)).T
Ar
```

```python
correcto(Ar)
```

(b) Crea una función llamada descomponer3x3 que reciba un array de enteros de 9×9 y lo separe en sus 9 subarreglos 3×3.

```python
def descomponer3x3(A):
    A = A.reshape(9,3,3)
    return A
```

```python
Ar1= np.eye(9,9)
descomponer3x3(Ar1)
```

```python
#sin embargo esta no es la descomposición que queremos para el sudoku ya que hace arreglos de 3 x 3 pero por filas. 
#Así que crearemos otra función que separe el arreglo como lo deseamos.
def Descomponer3x3(A):
    
    A1=A[:3,:3]
    A2=A[:3,3:6]
    A3=A[:3,6:]
    A4=A[3:6,:3]
    A5=A[3:6,3:6]
    A6=A[3:6,6:]
    A7=A[6:,:3]
    A8=A[6:,3:6]
    A9=A[6:,6:]
    
    Cubitos= np.array([A1,A2,A3,A4,A5,A6,A7,A8,A9])
    
    return Cubitos
```

(c) Crea la función esSolucion que recibe un array de 9×9 y devuelve si es o no solución del sudoku.

```python
def esSolucion(Array):
    #constantes   
    filas=0
    cols=0
    cubitos=0
    #nuevos arreglos
    Fila= Array.reshape(9,1,9)
    Columna= Array.reshape(9,9,1)
    Cubik= Descomponer3x3(Array)
    
    #comprobando para filas
    for m in range(0,9):
        if correcto(Fila[m])== "Completo":
            filas= filas+1
            
    #comprobando para columnas   
    for n in range(0,9):
        if correcto(Columna[n])=="Completo":
            cols= cols+1
            
    #comprobando para cubos de 3x3        
    for c in range(0,9):
        Cubikafila= Cubik.reshape(9,1,9)
        if correcto(Cubikafila[c])=="Completo":
            cubitos= cubitos+1
        
        
    if filas==9 and cols==9 and cubitos==9:
        return "Sí"
    else:
        return "No"
        
    
```
