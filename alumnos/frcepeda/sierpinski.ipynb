```python
def fib(n):
    a, b = 0, 1
    for i in range(n-1):
        a, b = b, a + b
    return b
```

```python
fib(10) # should be 55
```

```python
def factorial(n):
    f = 1
    for i in range(n):
        f = f * (i+1)
    return f
```

```python
factorial(5) # should be 120
```

```python
def SierpinskiTriangle(x1, y1, x2, y2, x3, y3, iterations = 5):
    if not iterations:
        plt.plot([x1, x2], [y1, y2], lw=1.0, color='black')
        plt.plot([x1, x3], [y1, y3], lw=1.0, color='black')
        plt.plot([x2, x3], [y2, y3], lw=1.0, color='black')
        return
    
    x12 = (x1 + x2) / 2.0
    x13 = (x1 + x3) / 2.0
    x23 = (x2 + x3) / 2.0
    y12 = (y1 + y2) / 2.0
    y13 = (y1 + y3) / 2.0
    y23 = (y2 + y3) / 2.0
    
    SierpinskiTriangle(x1, y1, x12, y12, x13, y13, iterations - 1)
    SierpinskiTriangle(x2, y2, x12, y12, x23, y23, iterations - 1)
    SierpinskiTriangle(x3, y3, x13, y13, x23, y23, iterations - 1)
```

```python
plt.figure(figsize = (6,6))
SierpinskiTriangle(0,0, 2,0, 1,math.sqrt(2))
```
