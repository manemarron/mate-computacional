```python
def factorial(n):
    if n == 0:
        return 1
    if n == 1:
        return 1
    else:
        return n * factorial(n-1)
```

```python
class CurvaVonKoch:
    def __init__(self, xi, yi,zi, xf, yf,zf):
        self.xi = xi
        self.yi = yi
        self.zi = zi
        
        self.xf = xf
        self.yf = yf
        self.zf = zf
        
    def _generar(self, xi, yi,zi, xf, yf,zf, iteracion):
        
        if iteracion == 0:
            self.dibujar(xi, yi,zi, xf, yf,zf)
            
        elif iteracion >= 0:
            x1i = (xi +yi) / 2.0
            y1i = (yi + zi) / 2.0
            z1i = (xi + zi)/2.0
            
            x1f = (xf +yf) / 2.0
            y1f = (yf + zf) / 2.0
            z1f = (xf + zf)/2.0       
            self._generar(xi, yi, zi, xf,yf,zf, iteracion - 1)
            #self._generar(x1i,y1i,z1i,x1f,y1f,z1f,  iteracion - 1)
            self._generar(x1i,yi,y1i,x1f,yf,y1f,  iteracion - 1)
            self._generar(z1i,y1i,zi,z1f,y1f,zf,  iteracion - 1)
            self._generar(xi,x1i,z1i,xf,x1f,z1f,  iteracion - 1)
            
            
    def dibujar(self, xi, yi, zi, xf, yf,zf):
        plot([xi, yi], [xf, yf], lw=1.0, color='black')
        plot([yi, zi], [yf, zf], lw=1.0, color='black')
        plot([xi, zi], [xf, zf], lw=1.0, color='black')
        
    def generar(self, iteraciones=0):
        return self._generar(self.xi, self.yi,self.zi, self.xf, self.yf,self.zf, iteraciones)
```

```python
c = CurvaVonKoch(0,4,2,0,0,2.36)
```
