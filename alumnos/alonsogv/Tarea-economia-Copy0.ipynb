```python
class Consumidor:
    
    def __init__(self, id, demandas = 0.0):
        self.id = id
        self.demandas = demandas
        
    def __repr__(self):
        return "<Consumidor %d> demandas: %d" % (self.id, self.demandas)
    
    def existe_necesidad(self):
        return self.demandas > 0
    
    def comprar(self, mercado):
        while self.existe_necesidad() and mercado.existe_oferta(): # Comprará mientras necesite satisfacer su demanda y haya oferta
            productor_mas_barato = mercado.productor_mas_barato()
            if productor_mas_barato: # Si hay un productor barato
                if productor_mas_barato.precio > PRECIO_MAXIMO_ACEPTABLE: # Si el precio es mayor que el aceptable
                    self.demandas = self.demandas * 0.5                   # reduzco mi demanda a la mitad

                oferta_mas_barata = productor_mas_barato.oferta
                
                if self.demandas > oferta_mas_barata:    # Si mi demanda supera a la oferta del productor
                    self.demandas = self.demandas - oferta_mas_barata #entonces mi demanda se ajuste a ser la diferencia entre 
                    productor_mas_barato.oferta = 0.0 #mi demanda actual y la oferta más barata. Y se agota el productor más barato.
                else:                                    # En este caso mi demanda es menor que la oferta del productor
                    productor_mas_barato.oferta = productor_mas_barato.oferta - self.demandas 
                    self.demandas = 0.0 #entonces el consumidor le compra únicamente lo necesario al productor, agotándose su demanda.
                    #Mientras que al productor solo le queda la diferencia.
```

```python
from functools import reduce
import math
import random
import numpy as np

class Mercado:
    
    def __init__(self, num_productores, num_consumidores, duracion):
        self.productores = []
        self.consumidores = [] #Constructor del mercado, que tiene como objetos, los productores, consumidores, la duración.
        self.demanda_generada = [] #Son de tipo array los consumidores, productores y la duración.
        
        self.duracion = duracion
        
        self.demanda_oferta = np.zeros([self.duracion, 2])
        self.precio_demanda = np.zeros([self.duracion, 2])
        
        for p_id in range(num_productores): #A cada productor se le asignan las condiciones de su constructor para inicializarlo.
            productor = Productor(id=p_id) #Supongo se refiere al id que identificará al producto. Producto 0, Productor 1, etc.
            productor.precio = COSTO + random.randint(0, GANANCIA_MAXIMA_INICIAL) #El precio del Producto i, es el COSTO (homogéneo para todos) más un entero arbitrario
            productor.oferta = random.randint(1,OFERTA_MAXIMA_INICIAL) #La oferta del productor, es un entero arbitrario.
            self.productores.append(productor)
            
        for c_id in range(num_consumidores): #Lo mismo para inicializar al consumidor.
            consumidor = Consumidor(id = c_id) #El id del consumidor (Consumidor 0, Consumidor 1, etc.)
            self.consumidores.append(consumidor) #Sabe Dios qué será esto.
            
        for t in range(duracion): #Duración.
            self.demanda_generada.append(round((math.sin(t)+2)*20))
    
    def precio_promedio(self):
        suma_precios = sum([productor.precio for productor in self.productores])
        return round(suma_precios/len(self.productores), 2)
        
    def oferta_total(self):
        return sum([productor.oferta for productor in self.productores])
        
    def demanda_total(self):
        return sum([consumidor.demandas for consumidor in self.consumidores])
        
    
    def existe_oferta(self):
        return self.oferta_total() > 0
    
    def existe_demanda(self):
        return self.demanda_total() > 0
        
    def productor_mas_barato(self):
        mas_barato = None
        productores = [ productor for productor in self.productores if productor.oferta > 0 ]
        if productores:
            mas_barato = min(productores) 
        return mas_barato
    
    def evolucionar(self, step):
        for consumidor in self.consumidores:
            consumidor.demandas = self.demanda_generada[step]    
            
        self.demanda_oferta[step] = [self.demanda_total(), self.oferta_total()]
            
        for productor in self.productores:
            productor.producir()
        
        self.precio_demanda[step] = [self.precio_promedio(), self.demanda_total()]
        
        while self.existe_demanda() and self.existe_oferta():
            for consumidor in self.consumidores:
                consumidor.comprar(mercado=self)
                
            for productor in self.productores:
                #Conforme se les va acabando el capital a los productores, por su incompetencia como vendedores, salen del mercado.
                if productor.capital < 0:
                    self.productores.remove(productor)
                
```

De las primeras intervenciones que nos enseñan en nuestras clases de Economía es el establecimiento de precios máximos y mínimos. Es decir, cuando el gobierno impone, para "ayudar al consumidor" un precio máximo; y análogamente, "para ayudar al productor" un precio mínimo. Esto con el objetivo de que ambos agentes queden protegidos. 
<br> Cuando sucede lo segundo (se establece un precio mínimo), sucede que hay un Exceso de Demanda (ver imagen). Provocando que el precio de equilibrio normal se re ajuste a la alza.
<br> En nuestro modelo, el gobierno elimina a los productores que no vendan con un precio mínimo. Y, en efecto, sucede lo esperado: el número de productores disminuye (obviamente), y el precio de intercambio se reajusta a la alta.
