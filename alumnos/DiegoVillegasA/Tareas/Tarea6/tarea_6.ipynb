## Problema 1

$$\textbf{J}(\beta_0, \beta_1) = \frac{1}{2n}\sum_{(x^{(i)}, y^{(i)}) \in X \times Y} (y^{(i)} - \hat{y}^{(i)}(x^{(i)}))^2 = \frac{1}{2n}\sum_{(x^{(i)}, y^{(i)}) \in X \times Y} (y^{(i)} - \beta_0 - \beta_1 x^{(i)})^2$$

```python
# Lo haré para m = 4
y0= sym.Symbol('y0')
y1= sym.Symbol('y1')
y2= sym.Symbol('y2')
y3= sym.Symbol('y3')
x0= sym.Symbol('x0')
x1= sym.Symbol('x1')
x2= sym.Symbol('x2')
x3= sym.Symbol('x3')
b0= sym.Symbol('b0')
b1= sym.Symbol('b1')

J = (1/8.0)*((y0-b0-b1*x0)**2+(y1-b0-b1*x1)**2+(y2-b0-b1*x2)**2+(y3-b0-b1*x3)**2)
```

**(a)** Usando **SimPy** demostrar que 

$$
\frac{\partial}{\partial \beta_j} \textbf{J}(\beta) = \frac{1}{m}\sum_{i=1}^{m}\left(\hat{y}(x^{(i)}) - y(x^{(i)})\right) \cdot x^{(i)}_j
$$

```python
Jdiff_b0 = sym.diff(J,b0)
Jdiff_b1 = sym.diff(J,b1)
```
