```python
def logistic_map(x,r):
    # Ejercicio
    return r*x*(1.0-x)
```

```python
from IPython.html.widgets import interact, fixed
from IPython.html import widgets
```

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plotMap(x,r):
    N=np.linspace(0,50,50)
    Y=np.zeros([50])
    Y[0]=x
    for i in range(49):
        Y[i+1]=logistic_map(x,r)
        x=Y[i+1]
    plt.scatter(N,Y)
```

```python
plotMap(0.35,2.8)
```

```python
plotMap(0.35,3.1)
```

```python
interact(plotMap, x=(0.0,2.0), r=(0.0,4.0))
```

```python
for r in [1.0,2.8,3.1,4.0]:
    X = np.linspace(0,1,50)
    C = logistic_map(X,r)
    plt.plot(X,C)
    plt.plot(X,X)
```
