### Integracion de Monte Carlo

```python
radio = 1

N = 10000

# Números al azar
X = np.random.uniform(low = -radio, high = radio, size  = N)
Y = np.random.uniform(low = -radio, high = radio, size = N)

# Distancia desde el centro
dist = np.sqrt(X**2+Y**2);

# Aplicamos la fórmula...
caja = (2.0*radio)**2
esta_adentro = dist < radio
N_adentro = np.sum(esta_adentro)
area_circulo = caja*N_adentro/N

# Dibujamos
plt.figure(figsize=(8,8))
plt.scatter(X,Y, c=esta_adentro, s=5.0, edgecolors='none', cmap=plt.cm.Paired)
plt.axis('equal')

print "[Monte Carlo] Área del círculo = ", area_circulo
print "[Exacta] Área del círculo = ", np.pi*radio**2
```

```python
y=[]
for i in (10,100,1000,10000,100000):
    radio = 1
    area_exacta=np.pi*radio**2
    
    N = i

    # Números al azar
    X = np.random.uniform(low = -radio, high = radio, size  = N)
    Y = np.random.uniform(low = -radio, high = radio, size = N)

    # Distancia desde el centro
    dist = np.sqrt(X**2+Y**2);

    # Aplicamos la fórmula...
    caja = (2.0*radio)**2
    esta_adentro = dist < radio
    N_adentro = np.sum(esta_adentro)
    area_circulo = caja*N_adentro/N
    
    error= abs(area_exacta-area_circulo)
    y.append(error)
    
x=[10,100,1000,10000,100000]
plt.figure(figsize=(8,8))
plt.scatter(x,y)

    
```

### Teorema del valor medio

```python
def f(x):
    return np.sin(x)

def intf(x):
    return -np.cos(x)
```

```python
a = 0
b = np.pi

N = 10000

X = np.random.uniform(low=a, high=b, size = N)
Y = f(X)

I_estimado = (b-a)*np.sum(Y) / N;

I_exacto = intf(b) - intf(a)

print "Estimación = ", I_estimado, "Exacto = ", I_exacto
```

### Ejercicio

```python
def f1(x):
    return (x**2)+4*x*(np.sin(x))

def intf1(x):
    return (x**3)/3+4*np.sin(x)-4*x*(np.cos(x))
```

```python
a=2
b=5*np.pi

N = 10000

X = np.random.uniform(low=a, high=b, size = N)
Y = f1(X)

I_estimado = (b-a)*np.sum(Y) / N;

I_exacto = intf1(b) - intf1(a)

print "Estimación =", I_estimado, " Exacto =", I_exacto, " Error =", np.abs(I_estimado-I_exacto)
```
