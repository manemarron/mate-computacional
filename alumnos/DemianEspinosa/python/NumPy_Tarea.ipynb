## Pregunta 2

En este ejercicio realizaremos una [**criba de Eratóstenes**](http://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

```python
def criba_eratostenes(maximo):
    bool_arr = np.ones((maximo,), dtype = bool)
    bool_arr[[0, 1]] = False #El 0 y 1 los hacemos aparte y los ponemos como no primos
    for i in range(2, int(np.sqrt(maximo)) + 1): #para todos los números entre 2 y la raiz entera del máximo
        if bool_arr[i]: #Sólo tachamos los múltiplos de los primos, porque si no lo son, ya se tacharon sus múltiplos antes.
            bool_arr[i * i :: i] = False #Una optimización es empezar en i² ya que i * algo_menor_a_i ya fue tachado antes. y apartir de ahí avanzar de i en i, para tomar los múltiplos de i.
    return bool_arr  
```

```python
for N in [10,100, 200]:
    print np.nonzero(criba_eratostenes(N))
```

## Pregunta 4

El **Sudoku** es un pasatiempo que consiste en un arreglo  $9 \times 9$ de celdas, con dígitos del 1 al 9, de tal manera que no se repiten valores en cada columna, cada renglón y en cada subarreglo de $3\times 3$.

(a) Crea una función llamada `correcto` que reciba un `array` $1d$ de tamaño 9 e indique si está completo o no (i.e. incluye los números del 1 al 9.

```python
def correcto(arr):
    assert arr.shape == (9,), "Arreglo inválido"
    arr = np.unique(arr)
    if arr.size < 9:
        return False
    for i in range (9):
        if arr[i] != i + 1:
            return False
    return True
```

(b) Crea una función llamada `descomponer3x3` que reciba un `array` de enteros de $9\times 9$ y lo separe en sus 9 subarreglos $3\times3$.

```python
def descomponer3x3(arr):
    assert arr.shape == (9,9), "Arreglo inválido"
    desc = np.zeros((9,9))
    for i in range(3):
        for j in range(3):
            for k in range(3):
                for l in range(3):
                    desc[i * 3 + j][k * 3 + l] = arr[i * 3 + k][j * 3 + l]
    return desc
```

```python
testoksudoku = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2], 
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],

    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],

    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
testwrongsudoku = np.array([range(i,10) + range(1,i) for i in range(1,10)])
#Casos de prueba sacados de la tarea de Freddy
```

```python
descomponer3x3(testoksudoku) 
```

(c) Crea la función `esSolucion` que recibe un `array` de $9\times 9$ y devuelve si es o no solución del sudoku.

```python
def esSolucion(arr):
    for i in range(9):
        if correcto(arr[i, : ]) == False:
            return False
    for i in range(9):
        if correcto(arr[:, i]) == False:
            return False
    for i in range(9):
        if correcto(descomponer3x3(arr)[i, :]) == False:
            return False
    return True
#Checa las líneas horizontales, verticales, y los subarreglos de 3x3.
```

```python
esSolucion(testoksudoku)
```

```python
esSolucion(testwrongsudoku)
```

## Pregunta 5

Utiliza el archivo `sunspots.txt` (localizado en la carpeta `data`), el cual contiene el número de manchas solares por cada mes desde enero de 1749. La primera columna es el mes, el segundo el número de manchas.

(a) Escribe el código que lea el archivo y haga una gráfica de las manchas en función del tiempo. Formatea el ejec $x$ para que se muestren cada 25 años empezando en $1750$. 

(b) Calcula el [promedio móvil](http://en.wikipedia.org/wiki/Moving_average). Dibújalo sobre el dibujo original.

```python
month, spots = np.loadtxt('../../../lecture_4/data/sunspots.txt', unpack = True, usecols = [0,1]) #Lee el archivo y lo guarda en month y spots

plt.figure(figsize = (20,5)) #Crea donde estará la gráfica

#Grafica meses contra spots colocando los labels descritos
plt.plot(month, spots, label = 'Sun spots')
plt.xlabel('Year')
plt.ylabel('Count')

plt.xlim(12, max(month)) 
ticks = xrange(12, int(max(month)), 25 * 12)
plt.xticks(ticks, [1749 + i/12 for i in ticks]) #pone las marcas empezando en 1750 y cada 25 años hasta el 2000

nmavg = 6 # Tomareos el promedio movil de los 6 valores a la izquierda y 6 a la derecha

mavg = np.zeros((len(month,)))
for i in range(len(month)):
    mavg[i,] = np.mean(spots[max(0,i-nmavg):min(len(spots),i+nmavg)])
plt.plot(month, mavg, color = 'red', linewidth = 2, label = 'Promedio Movil')

plt.legend(loc = 'upper left') #Crea el cuadro que dice que es cada una de los 2 plots
```
