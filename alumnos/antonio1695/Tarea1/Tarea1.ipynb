```python
"""Programa que muestra la suma de todos los multiplos de 3 y 5 para los numeros naturales entre 1 y mil.
Ejercicio 1 del Proyecto Euler. """
def multiplos(n):
    resultado = 0
    for i in(range(1,n+1)):
        if  i%3 == 0 or i%5 == 0:
            resultado = i + resultado
        


    return resultado

multiplos(1000)
            
```

```python
"""Programa que suma todos los numeros pares en la sucesion de Fibonacci (con límite el 4,000,000).
#Ejercicio 2 del Proyecto Euler."""
def fibonacci(n):
    if n ==0:
        return 0
    elif n==1:
        return 1
    else: 
        return fibonacci(n-1) + fibonacci(n-2)

def sumfib(n):
    resultado = 0
    for i in(range(1,n+1)):
        if fibonacci(i)% 2 == 0:
            resultado = resultado + fibonacci(i)
            
    return resultado

sumfib(4000000)
           
```

```python
"""Programa que obtiene el primo más grande de la factorización de primos de un número. 
Ejercicio 3 del Proyecto Euler."""

def factorizacion(n):
    resultado = n
    for i in(range(2,n)):
        while n%i==0:
            resultado = (resultado/i)
        if(resultado == 1):      
            return i
    
    
factorizacion(2)
```
