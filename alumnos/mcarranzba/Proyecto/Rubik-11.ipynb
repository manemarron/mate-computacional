## Una breve introducción a la Teoría de Grupos (o bien, al álgebra que usaremos en el proyecto)

No leer esta sección si el lector entiende los conceptos básicos de esta área de las matemáticas

**Definición: Grupo**
<br>
<div class="alert alert-info">
Sea $S \neq \emptyset $ un conjunto con una operación binaria definida "$*$". Se dice que la estructura algebraica $(S, *)$ es un grupo si satisface lo siguiente:
1. Si $(S, *)$ es asociativa.
2. Si existe $e$, neutro en $(S, *)$.
3. Si para todo elemento en $S$, existe un inverso en $(S, *)$

**Definición: Subgrupo**
<div class="alert alert-info">
Sea $(S,*)$ grupo, y $\emptyset \neq T \subseteq S $. Se dice que $T$ es un subgrupo de $S$ si $T$ es cerrado bajo "$*$", y si  $(T,*)$ es grupo.

**Definición: Producto cartesiano (generalización finita numerable)**
<div class="alert alert-info">
El producto cartesiano de un número finito de conjuntos $A_{1}, A_{2}, . . ., A_{n}$ es el conjunto de las $n$-tuplas, cuyo $k$-ésimo elemento pertenece a $A_{k}$, es decir: $$ $$ $A_{1} \times A_{2} \times . . . \times A_{n} = \{(a_{1}, a_{2}, . . ., a_{n}): a_{1} \in A_{1}, a_{2} \in A_{2}, . . . , a_{n} \in A_{n}\}$

#### Definición: Congruencia módulo $n$

Sea $n \in \mathbb N$, $n > 1$. Para $a,b \in \mathbb Z$, decimos que $a$ es congruente con $b$ módulo $n$, y escribimos $$a \equiv b \mod n,$$ si $a = b + kn$ para alguna $k \in \mathbb Z$

<div class="alert alert-info">

La congruencia módulo $n$ es una relación de equivalencia sobre $\mathbb Z$. Así, considerando los enteros que son congruentes módulo $n$, la clase de equivalencia de $a \in \mathbb Z$ vendría siendo: $$ $$ $[a] = \{ b \in \mathbb Z : a \equiv b \mod n \} $ $$ $$ Al conjunto de todas las clases de equivalencia, para la congruencia módulo $n$, la denotamos como $\mathbb Z_{n}$. $$ $$ Es decir, $\mathbb Z_{n} = \{ [a] : a \in \mathbb Z \}$, es el conjunto de todas las clases de equivalencia, de la relación de equivalencia congruente módulo $n$ $$ $$ 

**Proposición: Los enteros módulo $n$ forman un grupo**
<div class="alert alert-info">

Sea $n \in \mathbb N, n > 1 $. Definamos la operación binaria "$+_{n}$" en $\mathbb Z_{n}$: $$ $$ $[a] , [b] \in \mathbb Z_{n}$, $[a] +_{n} [b] = [a + b]$, (donde "$+$", representa la suma en los enteros). $$ $$ Luego, $(\mathbb Z_{n}, +_{n})$, es un grupo.

A partir de ahora, escribiremos "$+$" y "$a$", en lugar de "$+_{n}$" y "$[a]$", cuando en el contexto sea evidente que nos referimos a este grupo

**Definición: Permutación y grupos de permutaciones**

<div class="alert alert-info">

Una permutación es una función biyectiva. $$ $$ Sea $S$, conjunto no vacío, tal que $|S| = n \in \mathbb N$. Luego, el conjunto de funciones $S_{n} = \{f:S\to\ S: f$ es biyectiva $ \}$, junto con la operación binaria de composición de funciones "$\circ$", forman un grupo.

## Modelo del cubo físico

Las $26$ piezas que conforman el cubo se dividen en tres tipos: los centros (que sólo tienen un color asignado), las aristas (que tienen dos colores asignados), y los vertices (las esquinas, a quienes les corresponden tres colores). El espacio que una pieza ocupa, sin importar el tipo, se le llama cubículo. De esta forma, como las piezas centrales nunca cambian de cubículo, las usamos como punto de referencia para los cubículos que rodean cada una. Cuando uno realiza movimientos con el cubo, las piezas se mueven de cubículo en cubículo. Por lo anterior, podemos decir que el cubo está resuelto si las piezas están en su cubículo original, y si los colores de cada pieza están en su orientación correcta (de nuevo respecto al centro). $$ $$ Nuestro modelo consiste en una $4$-tupla $(\rho, v, \sigma, w)$, donde $\rho$  (respectivamente $\sigma$), describe cómo los vértices (respectivamente las aristas) están permutados, y $v$ (y $w$), describe cómo los vértices (y las aristas) están orientadas. Más aún, imaginemos que las piezas y los cubículos están enumerados. Cuando una pieza está en su posición natural, es porque su número asociado es igual al de su cubículo. $$ $$ Cualquier configuración del cubo arrojará dos permutaciones (ignorando las orientaciones de las piezas): $\rho \in S_{8}$, que corresponde a cómo los vértices han sido permutados; y $\sigma \in S_{12}$, que representa cómo las aristas fueron permutadas.

Para las orientaciones de cada pieza (la distribución de los colores), a su vez, enumeramos cada una de las caras o colores del cubo asociados a las piezas movibles, a saber, aristas y vértices (véase tabla). $0,1,2$ para los vértices (pues tienen tres colores), y $0,1$ para las aristas (pues tienen dos posibles colores). Por esta razón muy naturalmente trabajamos con $\mathbb Z_{3}$ y $\mathbb Z_{2}$, respectivamente. O más bien:
<br> <br> $w = (\ w_{1}, w_{2}, . . ., w_{12} )\ \in \mathbb Z_{2}^{12}$, es decir, $w = \{0, 1\}^{12}$, para la configuración de las $12$ aristas, donde $w_{i}$ corresponde al $i$-ésima arista  y el color que ésta contiene. 
<br> <br> $v = (\ v_{1}, v_{2}, . . ., v_{8} )\ \in \mathbb Z_{3}^{8}$, es decir, $v = \{0, 1, 2\}^{8}$, para la configuración de los $8$ vértices, donde $v_{j}$ corresponde a la $j$-ésimo vértice y el color que éste contiene.

Para simplificar la descripción de un cubo y poder ver rápidamente la posición relativa y sus orientaciones convenimos en nombrar las aristas y vértices de una forma única establecida por una convención arbitraria. Teniendo el cubo de frente comenzamos a nombrar los vértices. Este nombramiento se puede ver claramente en la tabla anexa. Se comienza por las aristas superiores y se recorre en el sentido de las manecillas del reloj. Después hacemos lo mismo con las inferiores. Para las aristas procedemos de igual manera: primero las $4$ de arriba, las $4$ de en medio y finalmente las $4$ de abajo. Además toda estampa tiene marcado lo que llamaremos número de orientación de la estampa que, para simplificar, se abreviará como NDO. Observamos en la imagen que las caras superior e inferior tienen puros ceros en sus NDO (tanto de los vértices como las aristas), y dos de las aristas tanto de la cara izquierda como la derecha. Esto nos será útil para identificar rápidamente en qué orientación se encuentra determinado vértice o arista.

Comenzaremos aplicando el método para las aristas. Habíamos dicho que las caras de arriba y abajo tienen puros ceros, en particular, tienen cero en sus estampas dentro de esquinas. Sus colores correspondientes (Verde y Azul) los llamaremos colores $0$; y su posición que ocupan cuando el cubo está resuelto la llamaremos posición $0$. Una pregunta pertinente sería: ¿qué representan los números $(0,1,2)$ en las orientaciones? La respuesta es simple, aunque puede ser difícil entenderlo al principio. *El NDO indica en caso de que un color $0$ estuviera aquí cuántas torciones en ésta esquina en sentido contrario a las manecillas del reloj (SCALMDR) necesitamos para que caiga en la posicion cero.

## Entrada válida

Ya estamos en condiciones de programar. Las siguientes funciones verifican que el vector de descripción de estado del cubo, en efecto, se trate de una entrada válida (es decir, que cumpla con la definición anterior).

```python
def Z2Valido(a): #Te dice si el arreglo solo tiene elementos de Z2
    resp=True
    for e in a:
        if e != 0 and e != 1:
            resp=False
    return resp

def Z3Valido(a): #Te dice si el arreglo solo tiene elementos de Z3
    resp=True
    for e in a:
        if e != 0 and e != 1 and e != 2:
            resp=False
    return resp

def EntradaValida(a,b,c,d): 

        """a array de tamaño 8, con elementos únicos por ser permutación
           b array de tamaño 8, cuyas entradas solo pueden ser 0,1,2
           b array de tamaño 12, con elementos únicos por ser permutación       
           d array de tamaño 12, cuyas entradas solo pueden ser 0,1"""

        EntradaValida = np.zeros(3).astype(bool)

        #Primero checamos que las longitudes sean válidas

        longitudes = np.zeros(4).astype(bool) #Crea un array de cuatro Falses. Se cambian las entradas por Trues si se cumple la condición.
        if len(a) == 8: #Permutación en S8
            longitudes[0] = True
        if len(b) == 8: #Elemento en (Z3)^8
            longitudes[1] = True
        if len(c) == 12: #Permutación en S12
            longitudes[2] = True
        if len(d) == 12: #Elemento en (Z2)^12
            longitudes[3] = True
        EntradaValida[0] = longitudes.all()  #Si todas las condiciones son válidas, devuelve True.  

        """Como a y b son permutaciones, no puede haber elementos repetidos.
        Para "modelarlas", la entrada i-ésima del array corresponde a la re ubicación del elemento a esa nueva posición,
        por ejemplo, (1,2,3,4,5,6,8,7), nos dice que se intercambió la entrada séptima por la octava.
        De esta manera, una permutación en S8 es válida si tiene ocho elementos únicos"""

        permutaciones = np.zeros(2).astype(bool)
        if len(np.unique(a)) == 8:
            permutaciones[0] = True
        if len(np.unique(c)) == 12:
            permutaciones[1] = True
        EntradaValida[1] = permutaciones.all() #Si todas las condiciones son válidas, devuelve True

        #Ahora hay que checar que b y d solo tengan 0,1,2 y 0,1 respectivamente.
        enterosModuloN = np.zeros(2).astype(bool) #Lo hacemos ahora con un arreglo inicialmente con Falses.

        if Z3Valido(b) == True:
            enterosModuloN[0] = True

        if Z2Valido(d) == True:
            enterosModuloN[1] = True

        EntradaValida[2] = enterosModuloN.all() #Si todas las condiciones son válidas, devuelve True

        return EntradaValida.all()
```

Olvidándonos un poco del modelo que hemos realizado de las configuraciones, es fácil hacer que los movimientos del cubo Rubik son un grupo, al que denotaremos, a falta de imaginación, como $(\mathbb G, \cdot)$. Aquí la operación "$\cdot$", para $R, L \in \{ X: X$ es un movimiento del cubo Rubik $\}$ $R \cdot L $ representa realizar la sucesión de movimientos $R$, y luego $L$ sobre el cubo.

<div class="alert alert-info">
**Proposición**: $\mathbb G = <D, U, R, L, F, B>$

Para dar el paso importante y llegar a los temas centrales que nos consciernen, primero revisemos una definición más:

**Definción: Acción de un grupo**

<div class="alert alert-info">
Una acción derecha de un grupo $(G,*)$ en un conjunto $A$ (no vacío), es una función $\circ :A \times G \to\ A$, tal que:

1. $(a \circ g_1) \circ g_2 = a \circ (g_1 * g_2)$, para toda $g_1, g_2 \in G$, y para toda $a \in A$
2. $a \circ e = a$, para toda $a \in A$, con $e$, neutro en $G$.

Cuando tenemos una acción de un grupo $G$ en un conjunto $A$, decimos simplemente que $G$ actúa en $A$.

Y por fin, el teorema o definición formal de $RC_3$:

**Teorema y definición: El Grupo del Cubo Rubik $RC_3$**

<div class="alert alert-warning">

El grupo $(\mathbb G,\cdot)$ actúa sobre el conjunto $RC_3^{*} = S_{8} \times \mathbb Z_{3}^{8} \times S_{12} \times \mathbb Z_{2}^{12}$, a través de los siguientes mapeos: 
<br> <br>$$D,\ U, \ R, \ L, \ F, \ B  : RC_3^{*} \times \mathbb G \to\ RC_3^{*}$$
 
<br> Donde para $x_0 = (\rho, (x_1,. . ., x_8),\tau, (y_1,. . ., y_{12})) \in RC_3^{*}$, arbitrario.
<br>
<br> $$x_0 \circ D = ((6\ 7 \ 8 \ 5)\rho,(x_{1},x_{2},x_{3},x_{4},x_{8},x_{5},x_{6},x_{7}), (10\ 11\ 12\ 9)\tau, (y_{1},y_{2},y_{3},y_{4},y_{5},y_{6},y_{7},y_{8},y_{12},y_{9},y_{10},y_{11}))$$ 
<br>
$$x_0 \circ U = ((4\ 3\ 2\ 1) \rho,(x_{2},x_{3},x_{4},x_{1},x_{5},x_{6},x_{7},x_{8}),(4\ 3\ 2\ 1)\tau,  
(y_{2},y_{3},y_{4},y_{1},y_{5},y_{6},y_{7},y_{8},y_{9},y_{10},y_{11},y_{12}))$$
<br>
$$x_0 \circ R = ((3\ 7\ 6\ 2)\rho,(x_{1},x_{6}+1,x_{2}+2,x_{4},x_{5},x_{7}+2,x_{3}+1,x_{8}),(7\ 10\ 6\ 2)\tau,(y_{1},y_{6}+1,y_{3},y_{4},y_{5},y_{10}+1,y_{2}+1,y_{8},y_{9},y_{7}+1,y_{11},y_{12}))$$
<br>
$$x_0 \circ L = ((5\ 8\ 4\ 1)\rho,(x_{4}+2,x_{2},x_{3},x_{8}+1,x_{1}+1,x_{6},x_{7},x_{5}+2),(5\ 12\ 8\ 4)\tau,
(y_{1},y_{2},y_{3},y_{8}+1,y_{4}+1,y_{6},y_{7},y_{12}+1,y_{9},y_{10},y_{11},y_{5}+1))$$
<br>
$$x_0 \circ F = ((4\ 8\ 7\ 3)\rho,(x_{1},x_{2},x_{7}+1,x_{3}+2,x_{5},x_{6},x_{8}+2,x_{4}+1), (8\ 11\ 7\ 3)\tau,
(y_{1},y_{2},y_{7},y_{4},y_{5},y_{6},y_{11},y_{3},y_{9},y_{10},y_{8},y_{12}))$$
<br>
$$x_0 \circ B = ((2\ 6\ 5\ 1)\rho,(x_{5}+1,x_{1}+2,x_{3},x_{4},x_{6}+2,x_{2}+1,x_{7},x_{8}),(6\ 9\ 5\ 1)\tau,
(y_{5},y_{2},y_{3},y_{4},y_{9},y_{1},y_{7},y_{8},y_{6},y_{10},y_{11},y_{12}))$$
<br>
<br> Luego, si tomamos a la identidad $\overrightarrow{\epsilon} = ( \epsilon_{8}, (0,. . .,0), \epsilon_{12}, (0, . . ., 0)\ ) \in RC_3^{*}$, el conjunto válido del cubo se define como:
$$RC_3 = \{ \overrightarrow{\epsilon} \circ M | M \in \mathbb G\}$$ Definimos la operación binaria "$*$" sobre $RC_3$ dada por:
<br> Para $y_0 = \overrightarrow{\epsilon} \circ M$, para alguna $M \in \mathbb G$
<br> Para $z_0 = \overrightarrow{\epsilon} \circ N$, para alguna $N \in \mathbb G$,
<br> $y_0*z_0 = \overrightarrow{\epsilon} \circ (M \cdot N)\in RC_3$

Unas cuantas observaciones antes de la prueba: en el teorema anterior se escribe la composición de las permutaciones como la simple concatenación de éstas, omitiendo el símbolo; además, éstas están escritas como ciclos (con la notación tal y como se vio en clase). Por último, recordemos que la sumas se realizan en congruente módulo $3$ o $2$, donde sea pertinente.
<br> Intutivamente lo que el teorema anterior nos dice es que cualquier vector de estado puede ser alcanzado a través de una sucesión finita de rotaciones válidas aplicadas al vector de estado inicial. Más aún, "multiplicar", dos vectores de estado válidos es equivalente a realizar en ese mismo orden ambas sucesiones de movimientos, partiendo del estado inicial.

**Demostación**
<br> Primero observemos que $\mathbb G = <D, U, R, L, F, B>$, por lo que solo es necesario establecer el mapeo con estos seis movimientos, y con ello decir que $\mathbb G$ actúa sobre todo el conjunto $RC_3^{*}$. 
<br> Consideremos una configuración arbitraria $x_0 = (\rho, \overrightarrow{x}, \tau, \overrightarrow{y}) \in RC_3^{*}$, y el movimiento $M_1 \in \mathbb G$. Al aplicárselo obtenemos una nueva configuración o vector de estado, digamos $x_0 \circ M_1 \in RC_3^{*}$. Si tenemos un nuevo movimiento de cubo, $M_2 \in \mathbb G$, y se vuelve a aplicar a la última configuración obtenida, obtendríamos una nueva: $(x_0 \circ M_1)\circ M_2 \in RC_3^{*}$, pero esto equivale a realizarle a la configuación $x_0$ el movimiento $(M_1\cdot M_2) \in RC_3^{*}$, es decir, $x_0 \circ (M_1\cdot M_2)\in RC_3^{*}$.
<br>Ahora, consideremos el movimiento nulo (la identidad en $\mathbb G$), al operarlo con cualquier vector de estado obtendríamos ese mismo vector, es decir: $x_0 \circ I = x_0$.
<br> Concluimos que $\circ$ es una acción de grupo.

Para probar que $RC_3$ es grupo, usaremos el siguiente teorema para dar una caracterización de todos los elementos que están en él.

## Cubo válido <br> ***Primer Teorema Fundamental del Cubo***

Claramente nuestro modelo tiene una mayor cantidad de configuraciones posibles que la que tiene en realidad un cubo físico.
Estamos tomando en cuenta también las posiciones ilegales del cubo: aquéllas que son imposibles de alcanzar con movimientos válidos (más adelante veremos un modelo para esto). Por ejemplo, es fácil ver que el vector de estado $( \epsilon_{8}, (0,. . .,0), \epsilon_{12}, (1, 0, . . ., 0)\ )$ ($\epsilon_{i}$ es la función identidad en $S_{i}$) es imposible de alcanzar, aunque sí cumple con la definición.
Pues justamente con el siguiente teorema podemos limitar las entraddas de los vectores de estado:

**Primer Teorema Fundamental del Cubo**

<div class="alert alert-warning">
Un vector de estado $(\rho, v, \sigma, w ) \in S_{8} \times \mathbb Z_{3}^{8} \times S_{12} \times \mathbb Z_{2}^{12} = RC_3^{*}$ representa una configuración legal del cubo si, y sólo si, satisface las siguientes propiedades:
<ul>
<li> $sign(\rho) = sign(\sigma) $
<li> $v_{1} + v_{2} + . . . + v_{8} \equiv 0$ $(mod$ $3) $
<li> $w_{1} + w_{2} + . . . + w_{12} \equiv 0$ $(mod$ $2) $

<br> Notación: $sign(.)$ se refiere a la paridad de la permutación

Donde "$sign$", para cada permutación anterior, se refiere a su paridad. Para conocer la definición de paridad de una permutación, ver: 
<br> http://en.wikipedia.org/wiki/Parity_of_a_permutation

**Demostración**

$\textbf{(i)}$ Primero tenemos que las tres condiciones son necesarias. Para esto vamos a probar que se satisfacen en el estado original (resuelto) del cubo, y que se mantienen bajo los $6$ movimientos básicos $R, L, U, D, F$ y $B$.
Cuando el cubo está resuelto le corresponde el vector $(\varepsilon,\overrightarrow{0},\varepsilon,\overrightarrow{0})$ y las tres condiciones se satisfacen. Cada permutación en los seis movimientos básicos es un ciclo de orden $4$ (par) y tiene signo $-1$. La suma de los componentes de cada vector de orientación de las esquinas es $0$ o $6$, que son divisibles entre $3$. La suma de los componentes de cada vector de orientación de aristas es $0$ o $4$, que es divisible entre $2$.

Si $X$ es una configuración legal con vector de posición $(\rho, v, \sigma, w )$ que satisface las tres condiciones, entonces después de aplicarle uno de los seis movimientos básicos, estas condiciones todavía se cumplen:
<li> La primera se cumple ya que cada movimiento causa al mismo tiempo ciclos de orden $4$ en los cubículos de las esquinas y las aristas, por lo tanto de misma paridad.
<li> La segunda se cumple porque $U$ y $D$ no cambian nada, mientras que $R,L,F$ y $B$ causan simultaneamente que dos componentes aumenten en $1(mod 3)$ y dos compenentes se reduzcan en $1(mod 3).$
<li> La tercera se cumple por la misma lógica. Los movimientos $U,D,F$ y $B$ no cambian nada, mientras que $R$ y $L$ incrementan y reducen dos componentes cada uno en $1(mod 2)$
Ya que toda configuración legal del cubo se obtiene aplicando estos movimientos al cubo resuelto, las tres propiedades se cumplen para cada configuración legal.

$\textbf{(ii)}$ Para probar que las tres condiciones son suficientes, tenemos que probar que cualquier vector de estado $(\rho, v, \sigma, w )$ que las satisface puede resolverse usando los seis movimientos básicos.
Usando el método por capas, se tiene ese resultado:

**Continúa demostración**
<br>
Sea $X$ una configuración correspondiente a $(\rho, v, \sigma, w )$.
Primero, sin pérdida de generalidad asumimos que $\rho$ y $\sigma$ tienen paridad igual a $1$ (en caso contrario solo aplicamos un movimiento de la cara y ya quedarían como queremos). Esto significa que la permutacion de esquinas es par, por lo que podríamos regresarlas a sus ubicaciones iniciales usando un ciclo de movimientos de orden 3. Análogamente para las aristas. Ya que podemos aplicar cualquier ciclo de orden 3 en esquinas o aristas, podemos regresar cualquier configuración a su posición original. Llamemos esta nueva configuración $X'$.
Como los movimientos básicos conservan las condiciones del primer teorema fundamental, el vector de estado $(\rho',v',\sigma',w')$ de $X'$ también las satisface y en este caso tenemos que $\rho'=\varepsilon, \sigma'=\varepsilon$.
Ahora falta mostrar que podemos voltear los "cubículos" a sus orientaciones correctas.

La tercera condición nos indica que tenemos que voltear un número par de aristas. Como tenemos series de movimientos para voltear cualquier par de aristas, entonces esta parte la podemos resolver.
Para las esquinas, la segunda condición nos dice que el número de vueltas en el sentido de las manecillas que tenemos que darles es igual al número de vueltas en sentido contrario módulo $3$. Así que primero volteamos cualquier par en los dos sentidos a sus orientaciones originales. El resultado va a ser que todas las vueltas que nos faltan van a ocurrir en tripletes del mismo sentido. Estos se pueden resolver usando una secuencia de movimientos de esquinas que ya conocemos.

Podemos concluir que $X$ es una configuración que podemos resolver. 

A merced del teorema anterior (la parte $\textbf{(i)}$, de la demostración), podemos caracterizar más fácilmente al conjunto de estados válidos del cubo: 
<br> $RC_3 = \{x \in RC_3^{*}: x$ cumple con el Teorema Fundamental del Cubo I $\}$, y que éste sea grupo con la operación proporcionada más arriba: las propiedades de grupo simplemente se "heredan" del grupo de movimientos del cubo, ya que $\mathbb G$ es asociativo, tiene inversos, y neutros.

Regresando al código, veamos cómo programar la entrada de datos y verificar que, en efecto, se trate de una entrada que pertenece a $RC_3$:

```python
def perm_parity(lst):
        '''\
        Dada una permutacion de digitos de 1..N en order como una lista, 
        regresa su paridad : +1 si es par; -1 si es impar.
        '''
        aux=[]
        for a in lst:
            aux.append(a)
        parity = 1
        for i in range(0,len(aux)-1):
            if aux[i] != i+1:
                parity *= -1
                mn = min(range(i,len(aux)), key=aux.__getitem__)
                aux[i],aux[mn] = aux[mn],aux[i]
        return parity 
    
def CuboValido(a,b,c,d): #La función devuelve True o False, dependiendo de si cumple la caracterización de RC3 o no.

    CV = np.zeros(4).astype(bool)
    #Condición 1:
    if np.sum(b) % 3 == 0:#la suma de las orientaciones de los vertices sea 0 modulo 3
        CV[0] = True
    #Condición 2:
    if np.sum(d) % 2 == 0:#la suma de las orientaciones de las aristas sea 0 modulo 2
        CV[1] = True
    #Condición 3:
    if perm_parity(a) == perm_parity(c):#checa que la paridad de las aristas y los vertices sea la misma
        CV[2] = True    
    #Condición extra: Entrada válida:
    if EntradaValida(a,b,c,d) == True:
        CV[3] = True
    return CV.all()
```

## Tipos de cubos que hay

Una caracteristica muy particular de la notación que escogimos para representar el cubo es que es relativamente sencillo contar el número de permutaciones que tiene. Para esto, nos servimos del orden de los vectores en los que se guarda la información y el Teorema Fundamental de Cubología o de cubos legales. El ejemplo más evidente es del cubo de $3\times 3\times 3$. 
Siendo $8!$ el orden de $S_{8}$, $3^8$ el de $Z_{3}^{8}$, $12!$ el orden de $S_{12}$ y $2^{12}$ el de $Z_{2}^{12}$. Además, tenemos por el $TFC$ que siendo... 

$X$ =$(\rho, v, \sigma, w ) \in S_{8} \times \mathbb Z_{3}^{8} \times S_{12} \times \mathbb Z_{2}^{12} $,
<ul>
<li> $sign(\rho) = sign(\sigma) $
<li> $v_{1} + v_{2} + . . . + v_{8} \equiv 0$ $(mod$ $3) $
<li> $w_{1} + w_{2} + . . . + w_{12} \equiv 0$ $(mod$ $2) $

Así, las últimas entradas de $v$ y $w$ quedan determinadas. Además al fijar la paridad $\rho$, solo la mitad de los $\sigma$ serán validos.

$$\Upsilon(3)=\frac{8! \cdot 3^8 \cdot 12! \cdot 2^12 }{3\cdot 2 \cdot 2}= 8! \cdot 3^7 \cdot 12! \cdot 2^{10} $$

```python
#Es un aleatorio usado para generar cadenas de orientaciones
def aleatorio(longitud,limiteinf, limitesup):#cuantos elementos tiene y hasta que enteros llega el aleatorio
    resp=[]
    for e in xrange(0,longitud,1):
        resp.append(random.randint(limiteinf, limitesup))
    return resp

#genera aleatoriamente una permutacion, se usa para aleatorizar posiciones
def permut(n):
    aux=[]
    resp=[]
    for e in xrange(1,n+1,1):
        aux.append(e)
    for i in xrange(0,n,1):
        #random.randint(0,n-1-e)
        resp.append(aux.pop(random.randint(0,n-1-i)))
    return resp

#genera aleatoriamente cadenas de rotaciones, las da en strings
def secuenciaAleatoria(numeroDeRotaciones):
    secuencia=aleatorio(numeroDeRotaciones,1,6)
    resp=""
    rotaciones = {#diccionario
            1 : 'D',
            2 : 'U',
            3 : 'R',
            4 : 'L',
            5 : 'F',
            6 : 'B',
            }

    for e in secuencia:
        resp=resp+rotaciones[e]
    return resp
    
#checa si dos arreglos son iguales término a término
def arreglosIguales(arr1,arr2):
    resp=True
    if(len(arr1)==len(arr2)):
        for e in xrange(0,len(arr1),1):
            if(arr1[e]!=arr2[e]):
                resp=False
    else:
        resp=False
    return resp
def deArregloaLista(arr1):
    lista=[]
    for i in xrange(0,len(arr1),1):
        lista.append(arr1[i])
    return lista
```

## El Algoritmo de Thistlethwaite

El siguiente método ha tenido muchas repercusiones importantes para la investigación y diseño de algoritmos de solución del cubo Rubik. Cabe destacar su importancia y belleza teórica que, por su propia naturaleza, lo invalidan para el uso humano, no así para las computadoras. Cuando el algoritmo ha sido mejorado e implementado con herramientas más sofisticadas de programación se puede lograr una solución en un promedio de $20$ movimientos.
<br> <p> La idea detrás del algoritmo es dividir el problema de resolución del cubo en problemas mucho más sencillos de completar (de nuevo, por una máquina). Para esto se va haciendo un "escalamiento" entre cada subgrupo, restringiendo cada vez más el número de movimientos permitidos. Los subgrupos son:
<div class="alert alert-info">
$G_{0} = \mathbb G =<L, R, F, B, U, D>$, todo el grupo del cubo está aquí.
<br> $G_{1} = <L, R, F^2, B^2, U, D>$, subgrupo de $G_{0}$,
<br> $G_{2} = <L^2, R^2, F^2, B^2, U, D> $, subgrupo de $ G_{1}$,
<br> $G_{3} = <L^2, R^2, F^2, B^2, U, D> $, subgrupo de $ G_{2}$,
<br> $G_{4} = \{e\}$, el cubo resuelto <br> <br> $G_{i}, 1 \leq i \leq 4$ corresponde a todos los estados del cubo que pueden ser alcanzados usando los movimientos en su conjunto generador. 

Intuitivamente lo que queremos hacer es ir maniobrando a lo largo de la cadena de subgrupos $G_{0} > G_{1} > G_{2} > G_{3} > G_{4}$. Supongamos un caso ideal en el que el rompecabezas fue bien encriptado (revuelto), y que solo podemos llegar a este estado usando todos los seis movimientos que generan a $G_{0}$. Para pasar a $G_{1}$, usamos los movimientos en el conjunto generador de $G_{0}$, a saber, $L, R, F, B, U, D$. Una vez que llegamos a $G_{1}$, usando únicamente movimientos en el conjunto generador de $G_{1}$ $(L, R, F^2, B^2, U, D)$, pasamos a $G_{2}$; y así, sucesivamente hasta llegar a $G_{4}$: el cubo resuelto.
<br> Ahora bien, para realizar la transición entre cada subgrupo, nos fijamos en las clases laterales. Por lo visto en clase, teniendo cualquier subgrupo de un grupo, las clases laterales forman una partición del grupo. Thistlethwaite primeramente diseñó unas tablas que representaran las clases laterales; y para cada elemento (cada clase lateral) en cada grupo cociente (o en cada tabla) halló la serie de movimientos que te llevarían dicho elemento al subgrupo más pequeño. <br> Lo anterior lo podríamos formalizar de la siguiente manera: 

**Algoritmo de Thistlethwaite**
<div class="alert alert-warning">
Sea $g_{0} \in G_{0}$, un estado bien encriptado del rompecabezas. 
<ul>
**Transición de $G_0$ a $G_1$**
<li> Ya que $G_1 < G_0$, podemos formar una partición de $G_0$, a partir de la unión de las clases laterales, es decir, $G_0 = \mathop{\bigcup}_{i = 1}^{r_{0}} g_{i,0} G_1$, con $g_{i,0} \in G_0$, para toda $1 \leq i \leq r_0$, para alguna $r_0 > 1$. 
<li> Luego, existe $x_{1} \in G_{0}$, tal que $g_0 \in x_1 G_1$, o bien, $g_0 = x_1 g_1$, para alguna $g_1 \in G_1$
<br> Se sigue que $g_1 = x_1^{-1} g_0 \in G_1$.
<br> Nótese que para llegar a $G_1$, solo se usan movimientos en $G_0$, pues $x_1^{-1} \in G_0$
<br> <br> **Transición de $G_1$ a $G_2$**
<li> Ya que $G_2 < G_1$, podemos formar una partición de $G_1$, a partir de la unión de las clases laterales, es decir, $G_1 = \mathop{\bigcup}_{i = 1}^{r_{1}} g_{i,1} G_2$, con $g_{i,1} \in G_1$, para toda $1 \leq i \leq r_1$, para alguna $r_1 > 1$. 
<li> Luego, existe $x_{2} \in G_{1}$, tal que $g_1 \in x_2 G_2$, o bien, $g_1 = x_2 g_2$, para alguna $g_2 \in G_2$
<br> Se sigue que $g_2 = x_2^{-1} g_1 = x_2^{-1} x_1^{-1} g_0 \in G_2$.
<br> De nueva cuenta, solo usamos movimientos en $G_1$, para llegar a $G_2$, puesto que $x_2^{-1} x_1^{-1} \in G_1$
<br> <br> **Transición de $G_2$ a $G_3$**
<li> Ya que $G_3 < G_2$, podemos formar una partición de $G_2$, a partir de la unión de las clases laterales, es decir, $G_2 = \mathop{\bigcup}_{i = 1}^{r_{2}} g_{i,2} G_3$, con $g_{i,2} \in G_2$, para toda $1 \leq i \leq r_2$, para alguna $r_2 > 1$. 
<li> Luego, existe $x_{3} \in G_{2}$, tal que $g_2 \in x_3 G_3$, o bien, $g_2 = x_3 g_3$, para alguna $g_3 \in G_3$
<br> Se sigue que $g_3 = x_3^{-1} g_2 = x_3^{-1} x_2^{-1} x_1^{-1} g_0 \in G_3$.
<br> <br> **Transición de $G_3$ a $G_4 = \{e\}$**
<li> Consideremos a $g_3^{-1}$, inverso de $g_3$. Simplemente, $e = g_3^{-1} g_3 \in G_4$. Por último, redefinimos: $x_4 = g_{3}$,
<br> y se sigue que $e = x_4^{-1} x_3^{-1} x_2^{-1} x_1^{-1} g_0$
