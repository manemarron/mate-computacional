```python
def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fib(n-1) + fib(n-2)
```

```python
def factorial(n):
    
    if n==0:
        return 1
    if n>=1:
        return n *factorial(n-1)
    
```

```python
factorial(5)
```

```python
class CurvaVonKoch:
    def __init__(self, xi, yi, xf, yf):
        self.xi = xi
        self.yi = yi
        self.xf = xf
        self.yf = yf
    
    def _generar(self, xi, yi, xf, yf, iteracion):
        
        if iteracion == 0:
            self.dibujar(xi, yi, xf, yf)
            
        elif iteracion >= 0:
            x1 = xi + (xf - xi) / 3.0
            y1 = yi + (yf - yi) / 3.0
            
            x3 = xf - (xf - xi) / 3.0
            y3 = yf - (yf - yi) / 3.0
            
            r = hypot(x3 - x1, y3 - y1) #da la longitud entre esos dos puntos
            alpha = np.arctan2((y3 - y1), (x3 - x1)) 
            alpha += np.pi/3
            
            x2 = x1 + r*np.cos(alpha)
            y2 = y1 + r*np.sin(alpha)
            
            
            self._generar(xi, yi, x1, y1, iteracion - 1)
            self._generar(x1, y1, x2, y2, iteracion - 1)
            self._generar(x2, y2, x3, y3, iteracion - 1)
            self._generar(x3, y3, xf, yf, iteracion - 1)
            
    def dibujar(self, xi, yi, xf, yf):
        plot([xi, xf], [yi, yf], lw=1.0, color='blue')
        
    def generar(self, iteraciones=0):
        return self._generar(self.xi, self.yi, self.xf, self.yf, iteraciones)
```

```python
c = CurvaVonKoch(0, 0, 9, 0)
```

```python
c.generar(iteraciones=1)
```

```python
class CopoVonKoch:
    def __init__(self, lado=3, iteraciones=1):
        self.lado = lado
        self.iteraciones = iteraciones
        
        self.x_vertice1 = 0
        self.y_vertice1 = 0
        
        self.x_vertice2 = lado * np.cos(2 * np.pi/3.0)
        self.y_vertice2 = lado * np.sin(2 * np.pi/3.0)
        
        self.x_vertice3 = lado * np.cos(np.pi / 3.0)
        self.y_vertice3 = lado * np.sin(np.pi / 3.0)
        
        self._generar()
        
    def _generar(self):
        c1 = CurvaVonKoch(self.x_vertice1, self.y_vertice1, self.x_vertice2, self.y_vertice2)
        c2 = CurvaVonKoch(self.x_vertice2, self.y_vertice2, self.x_vertice3, self.y_vertice3)
        c3 = CurvaVonKoch(self.x_vertice3, self.y_vertice3, self.x_vertice1, self.y_vertice1)
        
        c1.generar(self.iteraciones)
        c2.generar(self.iteraciones)
        c3.generar(self.iteraciones)
        
```

```python
class CurvaSierpinski:
    def __init__(self, xi, yi, xj, yj):
        self.xi = xi
        self.yi = yi

        self.xj = xj
        self.yj = yj
        
        alpha = np.pi/3
        dist = hypot((xj-xi),(yj-yi))

        xf = dist*np.cos(alpha)
        yf = dist*np.sin(alpha)
        
        self.xf = xf
        self.yf = yf
    
    def _generar(self, xi, yi, xj, yj, xf, yf, iteracion):
        
        
        if iteracion == 0:
            self.dibujar(xi, yi, xf, yf, xj, yj)
            
        elif iteracion >= 0:
            x1 =  (xf + xi) / 2.0
            y1 =  (yf + yi) / 2.0
            
            x2=  (xi + xj) / 2.0
            y2=  (yi + yj) / 2.0
            
            x3 =  (xf + xj) / 2.0
            y3 =  (yf + yj) / 2.0
                                              
            self._generar(xi, yi, xf, yf, xj, yj, iteracion - 1)
            self._generar(x1, y1, xf, yf, x3, y3, iteracion - 1)
            self._generar(x2, y2, x3, y3, xj, yj, iteracion - 1)
            self._generar(xi, yi, x1, y1, x2, y2, iteracion - 1)
                        
            
    def dibujar(self, xi, yi, xf, yf, xj, yj):
        plot([xi, xf], [yi, yf], lw=1.0, color='green')
        plot([xj, xf], [yj, yf], lw=1.0, color='red')
        plot([xi, xj], [yi, yj], lw=1.0, color='yellow')
        
    def generar(self, iteraciones=0):
        return self._generar(self.xi, self.yi, self.xj, self.yj, self.xf, self.yf, iteraciones)
```
