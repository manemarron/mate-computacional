```python
radio = 1

N = 100000

# Números al azar
X = np.random.uniform(low = -radio, high = radio, size  = N)
Y = np.random.uniform(low = -radio, high = radio, size = N)

# Distancia desde el centro
dist = np.sqrt(X**2+Y**2);

# Aplicamos la fórmula...
caja = (2.0*radio)**2
esta_adentro = dist < radio
N_adentro = np.sum(esta_adentro)
area_circulo = caja*N_adentro/N

# Dibujamos
plt.figure(figsize=(8,8))
plt.scatter(X,Y, c=esta_adentro, s=5.0, edgecolors='none', cmap=plt.cm.Paired)
plt.axis('equal')

print "[Monte Carlo] Área del círculo = ", area_circulo
print "[Exacta] Área del círculo = ", np.pi*radio**2
```

```python
def f(x):
    return x**2+4*x*np.sin(x)

def intf(x):
    return (1/3)*x**3+4*(np.sin(x) + 2*x**2*np.cos(x))
```

```python
a = 2
b = 5*np.pi

N = 1000000

X = np.random.uniform(low=a, high=b, size = N)
Y = f(X)

I_estimado = (b-a)*np.sum(Y) / N;

I_exacto = intf(b) - intf(a)

print "Estimación = ", I_estimado, "Exacto = ", I_exacto
```

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
plt.figure(figsize = (10,10))

num_particulas = 400

particulas = np.ones([num_particulas,2])*100  # Todas las partículas en el punto (100,100)


# Dibujamos la posición inicial
line, = plt.plot(particulas[:,0], particulas[:,1], 'ro')
plt.xlim(0,200)
plt.ylim(0,200)
```

```python
steps = 5000
for i in range(steps):
    for particula in range(num_particulas):
        particulas[particula, 0] += random.randint(-1, 1)
        particulas[particula, 1] += random.randint(-1, 1)
        
        # Verificar colisión
        x, y = (particulas[particula, 0], particulas[particula, 1])
        if x == 200:
            particulas[particula, 0] = 198
        elif x == 0:
            particulas[particula, 0] = 2
        
        if y == 200:
            particulas[particula, 1] = 198
        elif y == 0:
            particulas[particula, 1] = 2
            
```

```python
# Dibujamos la posición luego de N steps
line = plt.plot(particulas[:100,0], particulas[:100,1], 'ro')
plt.xlim(0,200)
plt.ylim(0,200)
```
