### Robot sin controlador

```python
r = Robot(
    km=0.006123, ke=0.006087, mp=1.13, mw=0.03, ip=0.0041, iw=0.000039,
    l=0.07, r=0.051, res=3, g=9.81
)
```

```python
def sinControl(n=5000,x=0,v=0,phi=0,phiv=0):
    dt = 0.0001
    initial_state = np.matrix([x, v, phi, phiv]).T
    hist = r.integrate(n,initial_state,dt)
    fig, ax = plt.subplots(2,2, figsize=(15,10))
    fig.suptitle("Sin Control ($\phi$=0)",fontsize=20)
    for i in range(2):
        for j in range(2):
            ax[i][j].plot(np.linspace(0,n*dt,n+1),hist[:,2*i + j])
            ax[i][j].set_xlabel("t(s)",fontsize=15)
    ax[0][0].set_ylabel("x(m)",fontsize=15)
    ax[0][1].set_ylabel("v(m/s)",fontsize=15)
    ax[1][0].set_ylabel("$\phi$(rad)",fontsize=15)
    ax[1][1].set_ylabel("vel. ang.(rad/s)",fontsize=15)
```

```python
interact(sinControl, n=(0,10000,5),x=fixed(0),v=(-0.5,0.5,0.05),phi=(-0.5,0.5,0.05),phiv=(-0.5,0.5,0.05))
```

```python
def controlador(n=1000, kp=0.15, ki=6, kd=0.0075):
    r2 = RobotWithController(
        km=0.006123, ke=0.006087, mp=1.13, mw=0.03, ip=0.0041, iw=0.000039,
        l=0.07, r=0.051, res=3, g=9.81, pid_kp=kp, pid_ki=ki, pid_kd=kd, pid_angle=0
    )
    dt = 0.001
    initial_state = np.matrix([0, 0.1, 0, 0]).T
    hist = r2.integrate(n,initial_state,dt)
    
    #plot
    fig, ax = plt.subplots(2,2, figsize=(15,10))
    fig.suptitle("Con PID ($\phi$=0)",fontsize=20)
    for i in range(2):
        for j in range(2):
            ax[i][j].plot(np.linspace(0,n*dt,n+1),hist[:,2*i + j],label="vi=0")
            ax[i][j].set_xlabel("t(s)",fontsize=15)
            ax[i][j].legend(loc="best")
    ax[0][0].set_ylabel("x(m)",fontsize=15)
    ax[0][1].set_ylabel("v(m/s)",fontsize=15)
    ax[1][0].set_ylabel("$\phi$(rad)",fontsize=15)
    ax[1][1].set_ylabel("vel. ang.(rad/s)",fontsize=15)
```

```python
interact(controlador,n=(500,2000,10),kp=(-1,1,0.001),ki=(-10,10,0.01),kd=(-5*10**-2,5**-2,10**-4))
```

```python
from matplotlib.patches import Rectangle
def controladorKalman(n=1000,kp=0.5, ki=1, kd=0.004):
    r2 = RobotWithController(
        km=0.006123, ke=0.006087, mp=1.13, mw=0.03, ip=0.0041, iw=0.000039,
        l=0.07, r=0.051, res=3, g=9.81, pid_kp=kp, pid_ki=ki, pid_kd=kd, pid_angle=0
    )
    dt = 0.001
    initial_state = np.matrix([0, 0, 0, 0]).T
    hist = r2.execute_kalman_filter(n,initial_state,dt)
    mean_angle = np.mean(hist[:,2])
    var_angle = np.var(hist[:,2])
    print mean_angle
    print var_angle
    #plot
    fig, ax = plt.subplots(2,2, figsize=(15,10))
    fig.suptitle("Con PID ($\phi$=0)",fontsize=20)
    for i in range(2):
        for j in range(2):
            ax[i][j].plot(np.linspace(0,n*dt,n+1),hist[:,2*i + j])
            ax[i][j].set_xlabel("t(s)",fontsize=15)
            extra = Rectangle((0, 0), 1, 1, fc="w", fill=False, edgecolor='none', linewidth=0)
            ax[i][j].legend([extra, extra, extra,extra], ("kp=%s"%kp, "ki=%s"%ki, "kd=%s"%kd, "$\phi$ promedio=%s"%mean_angle),loc="best")
    ax[0][0].set_ylabel("x(m)",fontsize=15)
    ax[0][1].set_ylabel("v(m/s)",fontsize=15)
    ax[1][0].set_ylabel("$\phi$(rad)",fontsize=15)
    ax[1][0].grid()
    ax[1][1].set_ylabel("vel. ang.(rad/s)",fontsize=15)
```
