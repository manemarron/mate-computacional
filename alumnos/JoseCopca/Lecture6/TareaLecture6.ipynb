Calcula el valor de Pi usando Metodos de Monte Carlo

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import numpy.random
```

### 2-esfera

```python
radio = 1

N = 100000

# Números al azar
X = np.random.uniform(low = -radio, high = radio, size  = N)
Y = np.random.uniform(low = -radio, high = radio, size = N)

# Distancia desde el centro
dist = np.sqrt(X**2+Y**2);

# Aplicamos la fórmula...
caja = (2.0*radio)**2
esta_adentro = dist < radio
N_adentro = np.sum(esta_adentro)
area_circulo = caja*N_adentro/N

# Dibujamos
plt.figure(figsize=(8,8))
plt.scatter(X,Y, c=esta_adentro, s=5.0, edgecolors='none', cmap=plt.cm.Paired)
plt.axis('equal')

print "[Monte Carlo] Área del círculo = ", area_circulo
print "[Exacta] Área del círculo = ", np.pi*radio**2
```

Usé el mismo código de la clase para integrar un circulo de radio uno y aproximar a pi

### 3-esfera

```python
radio = 1

N = 100000

# Números al azar
X = np.random.uniform(low = -radio, high = radio, size = N)
Y = np.random.uniform(low = -radio, high = radio, size = N)
Z = np.random.uniform(low = -radio, high = radio, size = N)

# Distancia desde el centro
dist = np.sqrt(X**2+Y**2+Z**2);

# Aplicamos la fórmula...
caja = (2.0*radio)**3
esta_adentro = dist < radio**2
N_adentro = np.sum(esta_adentro)
area_circulo = caja*N_adentro/N

# Dibujamos
plt.figure(figsize=(8,8))
plt.scatter(X,Y, c=esta_adentro, s=5.0, edgecolors='none', cmap=plt.cm.Paired)
plt.axis('equal')

print "[Monte Carlo] Área de la esfera = ", area_circulo
print "[Exacta] Área de la esfera = ", (4*np.pi*radio**3)/3
```

```python
radio = 1

N = 1000000

# Números al azar
X = np.random.uniform(low = -radio, high = radio, size = N)
Y = np.random.uniform(low = -radio, high = radio, size = N)
Z = np.random.uniform(low = -radio, high = radio, size = N)
W = np.random.uniform(low = -radio, high = radio, size = N)

# Distancia desde el centro
dist = np.sqrt(X**2+Y**2+Z**2+W**2);

# Aplicamos la fórmula...
caja = (2.0*radio)**4
esta_adentro = dist < radio**2
N_adentro = np.sum(esta_adentro)
area_circulo = caja*N_adentro/N

plt.figure(figsize=(8,8))
plt.scatter(X,Y, c=esta_adentro, s=5.0, edgecolors='none', cmap=plt.cm.Paired)
plt.axis('equal')

print "[Monte Carlo] Área de la 4-esfera = ", area_circulo
print "Aproximación de alfa=", (area_circulo)/(np.pi*radio**4)
```

El valor aproximado de alpha es 5/3
