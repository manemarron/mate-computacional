```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
```

```python
def plotDif(results):
    time = np.arange(len(results) +  1)
    dif = [0]
    for i in range(len(results)):
        if results[i] == 'T':
                dif.append(dif[i])
        elif results[i] == 'X':
                dif.append(dif[i] + 1)
        else:
                dif.append(dif[i] - 1) 
    
    plt.plot(time, dif)
    plt.xlabel('# Games')
    plt.ylabel('#A - #B')
    plt.title(u"Difference-Time Plot")
```

```python
plotDif('XXXOTXOTTTTTOXOOTOXOOOXOTTTTTTOXOOTOXOTOXOTOOOOOOOTTTTTTTTTX')
```

```python

def plotPie(results):
    A = B = T = 0
    for i in range(len(results)):
        if results[i] == 'T':
                T += 1
        elif results[i] == 'X':
                A += 1
        else:
                B += 1    
                
    #make a square figure and axes
    figure(1, figsize=(6,6))
    ax = axes([0.1, 0.1, 0.8, 0.8])

    # The slices will be ordered and plotted counter-clockwise.
    labels = 'Player A', 'Player B', 'Ties'
    fracs = [A, B, T]
    pie(fracs,labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
            
    title('Final A vs B', bbox={'facecolor':'0.8', 'pad':5})
    show()
```

```python
plotPie('XXXOTXOTTTTTOXOOTOXOOOXOTTTTTTOXOOTOXOTOXOTOOOOOOOTTTTTTTTTX')
```
