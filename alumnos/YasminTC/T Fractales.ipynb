```python
class Triangulo:
    def __init__(self, xa, ya, xb, yb, xc, yc):
        self.xa = xa
        self.ya = ya
        self.xb = xb
        self.yb = yb
        self.xc = xc
        self.yc = yc
        
    def _generar(self, xa, ya, xb, yb, xc, yc, iteracion):
        
        if iteracion == 0:
            self.dibujar(xa, ya, xb, yb, xc, yc)
            
        elif iteracion >= 0:
            self.dibujar(xa, ya, xb, yb, xc, yc)
            
            x1 = (xb + xa) / 2.0
            y1 = (yb + ya) / 2.0
            
            x3 = (xc + xa) / 2.0
            y3 = (yc + ya) / 2.0
            
            x2 = (xc + xb) / 2.0
            y2 = (yc + yb) / 2.0
            
            self._generar(xa, ya, x1, y1, x3, y3, iteracion - 1)
            self._generar(x1, y1, xb, yb, x2, y2, iteracion - 1)
            self._generar(x3, y3, x2, y2, xc, yc, iteracion - 1)
            
            
    def dibujar(self, xa, ya, xb, yb, xc, yc):
        plot([xa, xb], [ya, yb], lw=1.0, color='black')
        plot([xa, xc], [ya, yc], lw=1.0, color='black')
        plot([xb, xc], [yb, yc], lw=1.0, color='black')
        
    def generar(self, iteraciones=0):
        return self._generar(self.xa, self.ya, self.xb, self.yb, self.xc, self.yc, iteraciones)
```

```python
a=Triangulo(0,0,2,0,1,np.sqrt(3))
```
