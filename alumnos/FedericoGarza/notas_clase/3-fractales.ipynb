```python
def factorial(n):
    if n == 0:
        return 1
    return n*factorial(n-1)
```

```python
class sierpinski:
    
    def __init__(self, x1, x2, x3, y1, y2, y3):
        self.x1 = x1
        self.x2 = x2
        self.x3 = x3
        self.y1 = y1
        self.y2 = y2
        self.y3 = y3
        
    def _generar(self, x1, x2, x3, y1, y2, y3, iteracion):
        
        if iteracion == 0:
            self.dibujar(x1, x2, x3, y1, y2, y3)
        
        elif iteracion >= 0:
            x12 = (x1 + x2)/2.
            x13 = (x1 + x3)/2.
            x23 = (x2 + x3)/2.
            y12 = (y1 + y2)/2.
            y13 = (y1 + y3)/2.
            y23 = (y2 + y3)/2.

            self._generar(x1, x12, x13, y1, y12, y13, iteracion-1)
            self._generar(x12, x2, x23, y12, y2, y23, iteracion-1)
            self._generar(x13, x23, x3, y13, y23, y3, iteracion-1)
    
    def dibujar(self, x1, x2, x3, y1, y2, y3):
        plot([x1, x2, x3, x1], [y1, y2, y3, y1], lw = 1.0, color = 'black')
        
    def generar(self, iteraciones = 0):
        return self._generar(self.x1, self.x2, self.x3, self.y1, self.y2, self.y3, iteraciones)
        
```

```python
s = sierpinski(0., 1., .5, 0., 0., np.sqrt(1-0.5**2))
```
