### Programa Inversionista

Al aplicar la Regresión Lineal para los días que reducen el error se obtiene una predicción del precio de cierre

El programa inversionista consta de 2 inversionistas que:

    El primer inversionista (Inversionista Simple):
        Si la predicción es mayor al precio de cierre del día anterior, compra.
        Si la predicción es menor al precio de cierre del dia anterior, vende.
        
    El segundo inversionista (Inversionista Completo):
        Si la predicción es mayor al precio de cierre y el volumen aumentó los dos días anteriores, compra.
        Si la predicción es mayor al precio de cierre y el minimo más cercano es menor al anterior, compra.
        Si la predicción es menor al precio de cierre y el volumen aumentó los dos días anteriores, vende.
        Si la predicción es menor al precio de cierre y el mínimo más cercano es mayor al anterior, vende.

class SimulacionInversionSimple:
    
    def __init__(self, monto = 10000, dia_actual = 101):
        """

        self.stock = 0
        
    def RegresionLineal(self, x1, y):
        """
        x1 = vector de features para la regresión lineal
        y = vector de resultados para la regresión lineal
        """
        R = RegresionLinealUnivariada(alpha=0.00001, max_iters=1000, tols=0.0000001)
        self.vector = R.gradientDescent(x1[:self.dia_actual-1],y[:self.dia_actual-1])
        
    def DiaInversion(self, x1, y, vector):
        """
        Es el día en que se está invirtiendo :
            nos regresa el precio final y la regresión lineal
        x1 = vector de features para la regresión lineal
        y = vector de resultados para la regresión lineal
        v = vector de betas de la regresión lineal
        """
        self.resultados = [y[self.dia_actual], np.dot([1,x1[self.dia_actual]],vector)]
        return self.resultados
        
    def Inversionista(self, x1, y, vector):
        """
        El inversionista decide si vender o comprar basado en la regresión lineal :
            regresa el valor del portafolio a precio de cierre y el número de acciones en el
        x1 = vector de features para la regresión lineal
        y = vector de resultados para la regresión lineal

        if y[self.dia_actual-1] < self.resultados[1]:
            self.OrdenCompra(x1)
        elif self.resultados[1] < y[self.dia_actual-1]:
            self.OrdenVenta(x1)
        print 'El valor del portafolio es %s' % str(self.stock*self.resultados[0] + self.monto)
        print 'La cantidad de acciones en el portfolio es %s' % str(self.stock)
        return self.stock*self.resultados[0] + self.monto
            
    def OrdenCompra(self, x1):
        """
        La ejecución de la orden de compra:
            se compra si la regresión es mayor al precio de apertura
        x1 = vector con precios de apertura del día

        """

        self.monto = self.monto + precio
        self.stock = 0
        print 'Se vendieron las acciones a $ %s por acción' % str(x1[self.dia_actual])

class RegresionLinealUnivariada:

    def __init__(self, alpha=0.3, max_iters=100, tols=0.001):
        """
        Parámetros.
        ---------------
        alpha = Learning rate
        max_iters = Número máximo de iteraciones
        tols = definición de convergencia
        """
        self.alpha = alpha
        self.beta = np.zeros(2)
        self.max_iters = max_iters
        self.tols = tols
        self.breaking_iteration = None
        self.historia = {'costo':[], 'beta':[]}  # Con fines de graficación
        
    def gradientDescent(self, x, y):
        """
        Parámetros:
        ---------------
        x = vector de entrenamiento de features
        y = vector de entrenamiento de variable a predecir (target)
        """    
        
        # ajustamos el vector de features
        unos = np.ones((x.shape[0], 1))
        Xt = x.reshape(x.shape[0], 1)
        Xt = np.concatenate((unos, Xt), axis=1)
        
        i = 0
        prep_J = 0
        m, n = Xt.shape
        
        
        while i < self.max_iters:     
            # Actualizamos theta
            self.beta = self.beta - self.alpha * self.gradiente(Xt, y)
            
            J = self.costo(Xt, y)
            
            if abs(J - prep_J) <= self.tols:
                print 'La función convergió con beta: %s en la iteración %i' % ( str(self.beta), i )
                self.breaking_iteration = i
                break
            else:
                prep_J = J
            
            self.historia['costo'].append(J)
            self.historia['beta'].append(self.beta)                
            i += 1
        return self.beta
    
    def hipotesis(self, x):
        return np.dot(x, self.beta)
    
    def costo(self, x, y):
        m = x.shape[0]
        error = self.hipotesis(x) - y
        return np.dot(error.T, error) / (2 * m) 
    
    def gradiente(self, x, y):
        m = x.shape[0]
        error = self.hipotesis(x) - y        
        return np.dot(x.T, error) / m 
    
    def efectividad(self, x1, y, x11, yy, datos, bloque):

    def esperanzaerror(self, x1, y, datos):
        error = np.zeros(datos)
        vector = self.gradientDescent(x1[:datos], y[:datos])

            error[i] = abs(y[i] - np.dot([1,x1[i]],vector))

            error[i] = abs(y[i] - np.dot([1,x1[i]],vector))

class ReconocimientoTendencias:

    def __init__(self, x1, y, inicio = 0, fin = 500):
        """
        x1 = vector con el volumen de las acciones
        y = vector con los precios de cierre de las acciones
        inicio = el día inicial de la tendencia
        fin = el dia final de la tendencia
        """
        self.inicio = inicio
        self.fin = fin 
        
    def EncontrarMinimos(self, y, inicio, fin):
        """
        Encuentra los mínimos entre los dias inicio y fin:
            regresa el vector con los mínimos
        self.minimos es el vector de minimos locales encontrados
        """
        j = 0
        contador = self.contadormin(y, inicio, fin) - 1
        minimos = np.zeros(contador + 1)
        if y[inicio] < y[inicio + 1]:
            minimos[0] = y[inicio]
            j += 1
        for i in range(inicio + 1, fin - 1):
            if y[i] < y[i-1] and y[i] < y[i+1]:
                minimos[j] = y[i]
                j += 1
        if y[fin] < y[fin - 1]:
            minimos[contador] = y[fin]
        return minimos
    
    def EncontrarMaximos(self, y, inicio, fin):
        """
        Encuentra los máximos entre los dias inicio y fin:
            regresa el vector con los máximos
        self.maximos es el vector de máximos locales encontrados
        """
        j = 0
        contador = self.contadormax(y, inicio, fin) - 1
        maximos = np.zeros(contador + 1)
        if y[inicio] > y[inicio + 1]:
            maximos[0] = y[inicio]
            j += 1
        for i in range(inicio + 1, fin - 1):
            if y[i] > y[i-1] and y[i] > y[i+1]:
                maximos[j] = y[i]
                j += 1
        if y[fin] > y[fin - 1]:
            maximos[contador] = y[fin]
        return maximos
    
    def Tendenciamin(self, x1, y, inicio, fin):
        """
        Compara el minimo más cercano al día con el minimo anterior a ese:
            regresa True para comprar, False para no comprar
        x1 = vector de volumen
        y = vector de precios de cierre
        minimos = vector con los minimos locales
        sub_minimos = vector con los puntos de inflexión
        indice = dia en el que se encuetra el punto de inflexión
        """
        fin = fin - 1
        minimos = self.EncontrarMinimos(y, inicio, fin) 
        ultimo_minimo = minimos.shape[0] - 1
        sub_minimos = self.EncontrarMinimos(minimos,0,ultimo_minimo)
        ultimo_inflexion = sub_minimos.shape[0] - 1
        indice = self.buscaindice(ultimo_inflexion, y, inicio, fin)
        
        if ultimo_minimo < y[indice] and x1[fin-1] < x1[fin]:
            comprar = True
        else:
            comprar = False
        return comprar 
    
    def Tendenciamax(self, x1, y, inicio, fin):
        """
        Compara el máximo más cercano al día con el máximo anterior a ese:
            regresa True para vender, False para no vender
        x1 = vector de volumen
        y = vector de precios de cierre
        maximos = vector con los maximos locales
        sub_maximos = vector con los puntos de inflexión
        indice = dia en el que se encuetra el punto de inflexión
        """
        fin = fin - 1
        maximos = self.EncontrarMaximos(y, inicio, fin) 
        ultimo_maximo = maximos.shape[0] - 1
        sub_maximos = self.EncontrarMaximos(maximos,0,ultimo_maximo)
        ultimo_inflexion = sub_maximos.shape[0] - 1
        indice = self.buscaindice(ultimo_inflexion, y, inicio, fin)
        
        if ultimo_maximo > y[indice] and x1[fin-1] < x1[fin]:
            vender = True
        else:
            vender = False
        return vender
            
    def compara(self, x1, x2):
        """
        Compara los valores x2 y x1:
        """
        if x1 <= x2:
            return True
        elif x2 < x1:
            return False
        
    def buscaindice(self, valor, vector, inicio, fin):
        """
        Busca un valor en un vector y regresa la posición de dicho valor en el vector
        """
        indice = 0
        for i in range(inicio,fin):

                indice = i
                break
        return indice
